```python
headers = ["Date", "League", "Game", "Player", "Bet", "Line", "Pick", "Odds", "Bet Amount", "Units", "Payout","Result"]

myData = pd.DataFrame(columns = headers)

for row in data:
    myData.loc[len(myData)] = row

myData["Date"] = pd.to_datetime(myData["Date"], format="%m-%d-%Y")

myData
```

```python
def DataInput():
    prompt = input('Would you like to add bet data?')
    bet_list = []
   
    if prompt in ('Y','yes', 'Yes', 'y'):
        date_input = input('Date? (format: YYYY-MM-DD) ')
        
        date = datetime.strptime(date_input, '%Y-%m-%d')
        league = input('League? (Ex: NBA)')
        game = input('Game? (Ex: Nuggets v. Wizards)')
        player = input('Player? (Ex: Michael Jordan)')
        bet = input('Bet? (Ex: Points, ML)')
        line = float(input('Line? (Ex: 13.5)'))
        pick = input('Pick? (Ex: Over, Under') 
        odds = int(input('Odds (Ex: -135)'))
        bet_amount = float(input('Bet Amount? (Ex: 5.0)'))
        units = float(input('Units? (Ex: 1, 1.5)'))
        payout = float(input('Payout? (Ex: 12.0)'))
        results = input('Results? (Ex: Win, Loss)')
            
        bet_list.append([date, league, game, player, bet, line, pick, odds, bet_amount, units, payout, results])

        myData.loc[len(myData)] = bet_list
        
        print('Bet data added successfully!')
        
    else:
        return
 

DataInput()
```
