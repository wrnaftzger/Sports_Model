```python
from bs4 import BeautifulSoup
from selenium import webdriver
import requests
```

```python

driver = webdriver.Chrome() 
url = 'https://www.nba.com/stats/teams/traditional?Season=2023-24'
driver.get(url)

soup = BeautifulSoup(driver.page_source, 'html')
driver.quit()
```

```python
table = soup.find_all('table', class_="Crom_table__p1iZz")
print(table)
```

```python
titles = soup.find_all('th')[15:]
print(titles)
```

```python
headers2 = [title['title'] for title in titles if 'title' in title.attrs]  
headers1 = ['Rank','Team']
headers = headers1+headers2
print(headers) 
```

```python
import pandas as pd
```

```python
season23_24_df = pd.DataFrame(columns = headers)
season23_24_df.head()
```

```python
data = []
for table in table:  
    rows = table.find_all('tr')[1:]
    for row in rows:
        cells = row.find_all('td') 
        data.append([cell.text.strip() for cell in cells])  

print(data)
```

```python

for i in range(len(data)):
    season23_24_df.loc[i] = data[i]



season23_24_df
```
